```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('diabetic_data.csv')
data
```

## Priprema za klasterovanje

### Brisanje nedostajuÄ‡ih vrednosti, enkodiranje, feature selection

```python
bitni=['diag_1', 'diag_2', 'diag_3', 'max_glu_serum' ,'A1Cresult']


for b in bitni:
    most_frequent_value = data[b].mode().values[0]
    data[b] = data[b].replace(['?', 'None'], most_frequent_value)
```

```python
data.drop(['race','weight','payer_code','medical_specialty','encounter_id','patient_nbr', 'examide', 'citoglipton'], axis=1,inplace=True)
```

```python
feature_names = data.columns[0:]
feature_names
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
categorical_cols = data.select_dtypes(include=['object', 'category']).columns

for c in categorical_cols:
    data[c] = label_encoder.fit_transform(data[c])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

data_transformed = pca.fit_transform(data)
```

```python
X = data[feature_names]
X
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=feature_names)
```

```python
from sklearn.cluster import BisectingKMeans
```

```python
bkmeans = BisectingKMeans(n_clusters=2)
bkmeans
```

```python
bkmeans.fit(X)
```

```python
#koja instanca pripada kom klasteru
X[bkmeans.labels_ == 0]
```

```python
centers = pd.DataFrame(bkmeans.cluster_centers_, columns=data.columns)
centers
```

```python
# SSE
bkmeans.inertia_
```

```python
from sklearn.metrics import silhouette_samples

silhouette_values = silhouette_samples(X, bkmeans.labels_)
silhouette_values
```

```python
ks = range(2, X.shape[0]) # [)
inits = ['random', 'k-means++']
fig = plt.figure(figsize=(10,30))
idx = 1
silhouette = []
inertias = []
for k in ks:
    for init in inits:
        bkmeans = BisectingKMeans(n_clusters=k, init=init, n_init='auto')
        bkmeans.fit(X)
        if init == 'k-means++':
            inertias.append(bkmeans.inertia_)
            silhouette.append(silhouette_score(X, bkmeans.labels_))
```

```python
fig = plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(ks, inertias, marker='o')
plt.ylabel('inertia')
plt.xlabel('num_of_clusters')
plt.title('Elbow method')

plt.subplot(1, 2, 2)
plt.plot(ks, silhouette, marker="o")
plt.ylabel('silhouette coefs')
plt.xlabel('num_of_clusters')
plt.title('Silhouette coefs')

```
